Задание. Постройте график первой и второй производной функции: f(x) = sin(2x+1)**5


На отрезке [-5; 5].

```python
#Импортирую нужную библиотеку
import sympy as sp
```

```python
from sympy import *

x = symbols('x') #Превращаю x в символическое значение
expr = sin(2*x+1)**5 #Записываю функцию

#Использую метод lambdify() для превращения символического выражения в функцию
f = lambdify(x, expr, "math")
```

```python
expr
```

```python
res = sp.diff(expr, x) #Дифференцирую функцию по переменной x и нахожу первую производную функции
```

```python
res
```

```python
res2 = sp.diff(res, x) #Нахожу вторую производную функции
```

```python
res2
```

```python
#Строю график первой производной функции
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.1) #Указываю нужный отрезок, где находится x
y = 10*(np.sin(2*x+1)**4)*np.cos(2*x+1) #Вписываю функцию
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График первой производной функции')
plt.show()
```

```python
#Строю график второй производной функции по аналогии
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5, 5, 0.1)
y = (-20*np.sin(2*x+1)**5)+(80*np.sin(2*x+1)**3)*np.cos(2*x+1)**2
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('График второй производной функции')
plt.show()
```
